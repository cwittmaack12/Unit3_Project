- mean
> The average of all values.  
- median
> The middle number in the set.
- mode
> The number that appears most often in the data set.
- central tendency
> mean, median, and mode are all used to measure central tendency.  Single value that attempts to describe the central value 
in the data set.
- distribution
> Describes how the data spans the range of the data.  
- variance
> A value that describes how alike the data in the set is to the other data points.  measures variability from mean.
- standard deviation
> Derived from the variance.  Measures dispersion from the mean but it's measure in the same units as the data.

### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

<h3 style='color:red;'> First, load 'salary_numeric.csv' - where education has been re-expressed as numeric variables.</h3>

```python
# load salary_numeric.csv into dataframe where education levels have been re-expressed as numeric
df2=pd.read_csv('salary_numeric.csv')
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library

from scipy.stats import pearsonr, spearmanr 
```

### What is the correlation between salary and education?

#### Your Hypothesis = Strong positive correlation.  As education increased salary should also increase.

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(df2.salary, df2.education)
```

correlation of .477 indicates a moderate positive relationship between salary and education.

#### Are these results statistically significant? How do you know?

Yes.  The pvalue of 0 is less than the .05 threshold.

### What is the correlation between salary and age?

#### Your Hypothesis = I believe there is a strong positive correlations between salary and age.

```python
# calculate correlation and pvalue of salary and age with pearsonr method

pearsonr(df2.age,df2.salary)
```

Correlation of .663 indicates a strong positive correlation.

#### Are these results statistically significant? How do you know? 

Yes.  Since pvalue is less than .05 this would be considered statistically significant.

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = I believe there is a moderate correlation between salary and industry experience.

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
pearsonr(df2.salary, df2.industry_experience)
```

Moderate to strong correlation between salary and industry experience.

#### Are these results statistically significant? How do you know?

Yes, pvalue is less than .05.

### Heatmap of correlation matrix

```python
# create heatmap visualization of correlation matrix
plt.figure(figsize=(12,8))
g=sns.heatmap(df2.corr(), vmax=1, vmin=-1, cmap='BrBG_r', annot=True, square=True, center=0, linewidth=.85, 
            cbar_kws=dict(use_gridspec=False, location='top', shrink=.75) )
g.set_xticklabels(['Age', 'Education', 'Industry Experience', 'Salary'])
g.set_yticklabels(['Age', 'Education', 'Industry Experience', 'Salary'])
```

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
pd.crosstab(index=df2.industry, columns=df2.education, values=df2.salary, aggfunc=np.mean).round(0)
```

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# save above result to new variable
heat=pd.crosstab(index=df2.industry, columns=df2.education, values=df2.salary, aggfunc=np.mean).round(0)
heat
```

```python
# create heatmap with annotated mean salary in each industry and education level
plt.figure(figsize=(12,8))
g=sns.heatmap(heat, cmap='BrBG_r', annot=True, square=True, fmt='g', center=0, linewidth=.85, 
            cbar_kws=dict(use_gridspec=False, location='top', shrink=1) )
g.set_xticklabels(['High School', 'Some College', 'Bachelors', 'Masters'])
g.set_yticklabels(['Business', 'Finance', 'Healthcare', 'Tech'])
plt.xlabel('Education Level',fontsize=14)
plt.ylabel('Industry',fontsize=14)
plt.title('Heatmap of Salaries by Education Level Across Industries',fontsize=20)
```

##### Insights:

In general higher levels of education appear to have higher earning potential.  A bit more challenging to see by the colors but with numbers can also see that those in tech outearn their peers in other industries at all levels of education.
